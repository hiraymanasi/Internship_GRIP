## **Task 4 : To Explore Decision Tree Algorithm**


Our goal is to create the Decision Tree classifier for the given Iris dataset and visualize it graphically. So,if we feed any new data to this
classifier, it would be able to predict the right class accordingly.

```
#Importing libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
```

```
# Load the iris dataset
Iris = datasets.load_iris()
Iris_dframe = pd.DataFrame(Iris.data, columns = Iris.feature_names)
print(Iris_dframe.head(5))
```

```
#The target variable:
y=Iris.target
```

Let's split the dataset into a training set and a test set.


```
# Import train_test_split function
from sklearn.model_selection import train_test_split 
X=Iris_dframe
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

We have split the data as 70% training and 30% test set.

Let's create a Decision Tree Model.

```
# Importing Decision Tree Classifier , scikit-learn metrics module
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics 
```

```
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

We trained our algorithm. Now lets check the accuracy by comparing actual test set values and predicted values.

```
##Accuracy
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

Here, we got a classification rate of 95.55%, which is good accuracy.

Let us visualize the Decision Tree.

```
#need to install graphviz and pydotplus.
!pip install graphviz
!pip install pydotplus
```

```
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
import pydotplus

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = Iris.feature_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Iris.png')
Image(graph.create_png())
```

Let's see how changing of the criteria affects the decision tree .

```
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3) #Entropy is the criteria.

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

Well, there is no change in accuracy of the model.

Lets visulize the decision tree.

```
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = Iris.feature_names)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Iris.png')
Image(graph.create_png())
```

Now we have trained our algorithm so that giving new data it will predict the class with accuracy 95%
